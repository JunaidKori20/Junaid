```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fileN.csv')
```

```python
df
```

```python
print(df.head())
```

```python
print(df.describe(include='all'))
```

```python
print(df.isnull().sum())
```

```python
filtered_data = df[df['type'] == 'Movie']
print(filtered_data)
```

```python
df['country'] = df['country'].fillna('Unknown')
print(df['country'])
```

```python
df['description_length'] = df['description'].str.len()
print(df[['description', 'description_length']])
df
```

```python
df = df.drop(columns=['cast'])
print(df.head())
```

```python
release_year_array = df['release_year'].values
print(release_year_array)
```

```python
mean_release_year = np.mean(df['release_year'].dropna())
print(mean_release_year)
```

```python
type_counts = df.groupby('type').size()
print(type_counts)
```

```python
median_release_year = np.median(df['release_year'].dropna())
print(median_release_year)
```

```python
std_release_year = np.std(df['release_year'].dropna())
print(std_release_year)
```

```python
unique_ratings = np.unique(df['rating'].dropna())
print(unique_ratings)
```

```python
type_counts = df['type'].value_counts()
plt.bar(type_counts.index, type_counts.values, color='skyblue')
plt.title('Count of Types (Movies vs TV Shows)')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['release_year'].dropna(), bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
rating_counts = df['rating'].value_counts()
plt.pie(rating_counts.values, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings')
plt.show()
```

```python
df['duration_minutes'] = df['duration'].str.extract('(\d+)').astype(float)
plt.scatter(df['release_year'], df['duration_minutes'], alpha=0.5, color='purple')
plt.title('Duration vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.show()
```
